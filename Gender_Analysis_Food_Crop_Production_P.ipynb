# Gender Analysis of Food Crop Production Among Smallholder Farmers

**Synthetic portfolio project (120 respondents).**

## Introduction
This notebook demonstrates exploratory data analysis and hypothesis testing using a synthetic dataset.

## Objectives
- Compare male and female farmers
- Visualize production
- Perform an independent samples t-test

```
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

df=pd.read_csv('food_crop_data.csv')
df.head()
```

```
df.info()
df.describe(include='all')
```

```
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()
```

```
df.groupby('Gender')['Production_Output_tons'].mean().plot(kind='bar')
plt.title('Average Production Output by Gender')
plt.show()
```

```
male=df[df.Gender=='Male']['Production_Output_tons']
female=df[df.Gender=='Female']['Production_Output_tons']
print(ttest_ind(female,male))
```
