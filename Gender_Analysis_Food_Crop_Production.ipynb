# Gender Analysis of Food Crop Production
Synthetic dataset of 120 respondents.


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
df=pd.read_csv('../data/food_crop_data.csv')
df.head()
```

```python
print(df.groupby('Gender')['Production_Output_tons'].mean())
df.groupby('Gender')['Production_Output_tons'].mean().plot(kind='bar')
plt.show()
```

```python
m=df[df.Gender=='Male']['Production_Output_tons']
f=df[df.Gender=='Female']['Production_Output_tons']
print(ttest_ind(f,m))
```
